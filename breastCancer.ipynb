```python
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential
```

```python
X = pd.read_csv("data/entradas_breast.csv")
```

```python
Y = pd.read_csv("data/saidas_breast.csv")
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25)
```

```python
X_train.shape, Y_train.shape
```

```python
X_test.shape, Y_test.shape
```

```python
# 30 + 1 / 2
neural_network = Sequential([
    tf.keras.layers.InputLayer(shape = (30,)),
    tf.keras.layers.Dense(units = 16, activation = 'relu', kernel_initializer = 'random_uniform'),
    tf.keras.layers.Dense(units = 1, activation = 'sigmoid')
])
```

```python
neural_network.summary()
```

```python
neural_network.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['binary_accuracy'])
```

```python
neural_network.fit(X_train, Y_train, batch_size=10, epochs = 100)
```

```python
 predictions = neural_network.predict(X_test)
predictions = predictions > 0.7
```

```python
predictions
```

```python
Y_test
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(Y_test, predictions)
```

```python
confusion_matrix(Y_test, predictions)
```
